```python
#if we want to do a barplot for every categorical variable?
#can we create a function?
def categorical_barplot(variable):
    #barplot
    #the index is 1 or 0, and the values are the repetitions
    plt.bar(data[variable].value_counts().index,data[variable].value_counts())
    plt.xticks(data[variable].value_counts().index, data[variable].value_counts().index.values)
    plt.ylabel("Count of persons")
    plt.xlabel(variable)
    plt.show()
```

```python
categorical_barplot("sex")
#we can see the number of mens is greather than the number of womans
```

```python
#We want to define a pie chart, which allows us
#in the future to make different pie plots for each
#categorical variable that only can take 2 values.
def pie_chart(v_categorical,v0,v1):
    sizes = [data[v_categorical].value_counts()[v0], 
            data[v_categorical].value_counts()[v1]]
    labels = [v0,v1]
    fig, ax = plt.subplots()
    ax.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90)
    ax.axis("equal")
    plt.legend()
    plt.show()
```

```python
pie_chart("output",0,1)
#we can see thar 55% of persons had suffered a 
#heart attach
```

```python
#if you want to add more divisions of the pie chart
#is the same funcion but you need to add more arguments
def pie_chart_1(v_categorical,v0,v1,v2):
    sizes = [data[v_categorical].value_counts()[v0], 
            data[v_categorical].value_counts()[v1],
            data[v_categorical].value_counts()[v2]]
    labels = [v0,v1,v2]
    fig, ax = plt.subplots()
    ax.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90)
    ax.axis("equal")
    plt.legend()
    plt.show()
```

```python
pie_chart_1("slp",0,1,2)
```

```python
#can we make multiple pie chart for every categorical variable?
#the functions helps us te be more fast, but we can do the plot
#in a best wey, in terms of efficency

#1. First, we need to define which variables can define as categorical
categorical_variables = ["sex","cp","fbs","restecg",
                        "exng","slp","caa","thall",
                        "output"]
#2. Create the loop
n=0
for variable in categorical_variables:
    #gives me the index
    labels = data[variable].value_counts().index
    #gives me the count of unique values
    sizes = data[variable].value_counts().values
    #3x3 plot and n+1 is adding one plot
    plt.subplot(3,3,n+1)
    plt.pie(sizes,labels=labels,startangle=50,textprops={'fontsize': 6},
           autopct='%1.1f%%')
    plt.title(f'% of {variable}',color = 'black',fontsize = 8)
    n+=1
```

```python
#now, we want to analyze each variable which is not categorical
#for this reazon is necesary to create a function that shows
#the distribution of each variable in the two cases:
#if the person has had a heart attack or not
def not_categorical_plot(variable,categorical):
    #when i use fig ax, fig ax return a tuple that contains
    #a figure and objects
    fig, ax = plt.subplots()
    ax.hist(data[data[categorical]==1][variable],alpha=0.5,label="1",color="red",bins=20)
    ax.hist(data[data[categorical]==0][variable],alpha=0.3,label="0",color="blue",bins=20)
    plt.ylabel("Frecuency")
    plt.xlabel(variable)
    plt.legend()
```

1Accuracy-numbers of correct predictions mad by model divided by the total number of predictions. Accuracy is not a good choice with unbalanced classes (or the choice unbanced data) 2Recall: ability of a model to find all the relevant cases within a dataset. The precise definition of recall is the number of true positives divided by the number of true positives plus the numbers of false negatives. 3Precision, ability of a clasification model to identify only the relevant data points. Precision is definided as the number of true positives divided by the number of true positive plus the number of false positives

F1 = 2*(precision*recall)/(precision+recall) we use this instead mean, because this formula punishes extreme values
